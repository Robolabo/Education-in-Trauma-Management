```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1=('yes','no')
y1=(7,35)
fig1=plt.bar(x1,y1,width=0.8,alpha=0.8,label='Students')
plt.title('Medical Students')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
x2=('yes','no','not stated')
y2=(22,16,4)
fig2=plt.bar(x2,y2,width=0.8,alpha=0.8,label='Mannequin', color='orange')
plt.title('Use of high fidelity mannequin')
plt.legend()
plt.tight_layout()
plt.show()
```
